```python
import statistics
import numpy as np
import scipy.stats
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# The use of natural family planning remains small, but has seen growth over the last decade
# We've also seen that the overall number of women receiving contraception prescriptions has declined rapidly

over_time = pd.read_csv("contraception_over_time.csv", sep = ",")

# over_time.head()
# over_time.describe()
# over_time.info()

mean_ = statistics.mean(over_time['Natural family planning'])
print('The mean % of population using natural family planning over time is: ')
print(mean_)

x = over_time['Year']
y1 = over_time['Natural family planning']
y2 = over_time['Contraception Prescriptions (000s)']

#define subplots
fig,ax = plt.subplots()

#add first line to plot
ax.plot(x, y1, color="teal")

#add x-axis label
ax.set_xlabel('Year', fontsize=14)

#add y-axis label
ax.set_ylabel('% Using Natural Family Planning', color="teal", fontsize=16)

#define second y-axis that shares x-axis with current plot
ax2 = ax.twinx()

#add second line to plot
ax2.plot(x, y2, color="magenta")

#add second y-axis label
ax2.set_ylabel('Contraception Prescriptions (000)s', color="magenta", fontsize=16)

plt.title('Usage of Natural Family Planning and Emergency Contraception Over Time', fontsize=10)
```

```python
# Nat family planning is the only method of contraception to have grown, with oral contraceptives declining in popularity

x = over_time['Year']

y1 = over_time['Natural family planning']
y2 = over_time['Emergency Contraception Provided (000s)']
y3 = over_time['LARCs total']
y4 = over_time['Oral contraceptives']

plt.plot(x,y1,'g--',label='Natural Family Planning', linewidth=2)
plt.plot(x,y2,'lightblue',label='Emergency Contraception',linewidth=2)
plt.plot(x,y3,'b',label='Larcs',linewidth=2)
plt.plot(x,y4,'darkblue',label='Oral Contraceptives',linewidth=2)

plt.title('Correlation of contraceptive methods', fontsize=14)
plt.ylabel('Number of females using (000s)', fontsize=14)
plt.xlabel('Years', fontsize=14)

plt.legend(loc='upper right', fontsize=10)
```

```python
# The necessity of alternative methods surged during the pandemic; apps fulfilled this need

x = over_time['Year']

y1 = over_time['% of contacts at a health centre']
y2 = over_time['% of contacts in educational premises']
y3 = over_time['% of contacts at home']
y4 = over_time['% of contacts elsewhere']


plt.plot(x,y1,'g--',label='Health Centre', linewidth=2)
plt.plot(x,y2,'lightblue',label='Educational Premise',linewidth=2)
plt.plot(x,y3,'b',label='Home',linewidth=2)
plt.plot(x,y4,'darkblue',label='Elsewhere',linewidth=2)

plt.title('Places of access to Sexual Health Advice', fontsize=14)
plt.ylabel('% of females', fontsize=14)
plt.xlabel('Years', fontsize=14)

plt.legend(loc='center left', fontsize=10)
```
